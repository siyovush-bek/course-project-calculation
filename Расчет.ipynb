```python
# количество элементов профиля
m = 17

# количество клетей
n = 11
```

### Толщина загатовки

```python
s = 2
```

## Расчет

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("исходные_данные.csv")
L_n = data['L'].to_numpy()
R_n = data['R'].to_numpy()
A_n = np.radians(data['A'].to_numpy())
angles = pd.read_csv("матрица_углов.csv", header=None)
coeff_matrix = pd.read_csv("матрица_коэффициентов.csv", header=None)
R_n * np.abs(A_n)
L_n += R_n * np.abs(A_n)
```

### Размеры конечнего сечения

```python
pd.DataFrame({'Ln':L_n, 'An':A_n, 'Rn':R_n}, index=np.arange(1, 18))
```

### Сумма длин элементов профиля

```python
b0 = np.sum(L_n)
b0
```

### Матрица углов A

```python
A = angles.to_numpy()
A = np.abs(np.radians(A))
A_df = pd.DataFrame(A, index=np.arange(1, 18), columns=np.arange(1, 12))
A_df
```

### Матрица радиусов R

```python
R = np.zeros(shape=(m,n))

for i in range(1, 3):
    for j in range(m):
        if A[j, i] == 0:
            R[j, i] = 0
        else:
            R[j, i] = L_n[j] / np.abs(A[j, i])

for i in range(3, n):
    for j in range(m):
        if A[j, n-1] == 0:
            R[j, i] = 0
        else:
            R[j, i] = L_n[j] / np.abs(A[j, n-1])
```

```python
R_df = pd.DataFrame(R, index=np.arange(1, 18), columns=np.arange(1, 12))
R_df
```

### Матрица коэффициентов M

```python
M = coeff_matrix.to_numpy()
pd.DataFrame(M, index=np.arange(1, 18), columns=np.arange(1, 18))
```

### Матрица свободных коэффициентов b

```python
def b(j):
    if j < 0 or j >= m:
        raise ValueError
    return np.array([
        L_n[0] + L_n[1],
        R[1, n-1] * A[1, j],
        L_n[2] + L_n[3],
        R[3, n-1] * A[3, j],
        L_n[4] + L_n[5],
        R[5, j] * A[5, j],
        L_n[6] + L_n[7],
        R[7, j] * A[7, j],
        L_n[8],
        R[9, j] * A[9, j],
        L_n[9] + L_n[10],
        R[11, j] * A[11, j],
        L_n[11] + L_n[12],
        R[13, n-1] * A[13, j],
        L_n[13] + L_n[14],
        R[15, n-1] * A[15, j],
        L_n[15] + L_n[16]
    ])
```

### Матрица длин L

```python
L = [np.linalg.solve(M, b(i)) for i in range(n)]
L = np.column_stack(L)
```

```python
L_df = pd.DataFrame(L, index=np.arange(1, 18), columns=np.arange(1, 12))
L_df
```

### Экспорт данных в формат excel

```python
from openpyxl import Workbook
```

```python
wb = Workbook()

sheet = wb.active
sheet.cell(row=1, column=1).value = "s"
sheet.cell(row=1, column=2).value = s
sheet.cell(row=1, column=3).value = "мм"

row = 2

for j in range(n):
    for i in range(m):
        sheet.cell(row=row, column=1).value = f"L{i+1}_{j+1}"
        sheet.cell(row=row, column=2).value = L[i,j]
        sheet.cell(row=row, column=3).value = "мм"
        row += 1
        sheet.cell(row=row, column=1).value = f"R{i+1}_{j+1}"
        sheet.cell(row=row, column=2).value = R[i,j]
        sheet.cell(row=row, column=3).value = "мм"
        row += 1
        sheet.cell(row=row, column=1).value = f"A{i+1}_{j+1}"
        sheet.cell(row=row, column=2).value = A[i,j]
        sheet.cell(row=row, column=3).value = "град"
        row += 1
wb.save(filename="результат.xlsx")
```
